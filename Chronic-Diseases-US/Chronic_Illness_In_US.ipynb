The CPS uses a set of six questions to identify persons with disabilities. A response of “yes” to any one of the questions indicates that the person in question has a disability.

Is anyone deaf or does anyone have serious difficulty hearing? Is anyone blind or does anyone have serious difficulty seeing even when wearing glasses? Because of a physical, mental, or emotional condition, does anyone have serious difficulty concentrating, remembering, or making decisions? Does anyone have serious difficulty walking or climbing stairs? Does anyone have difficulty dressing or bathing? Because of a physical, mental, or emotional condition, does anyone have difficulty doing errands alone such as visiting a doctor's office or shopping?

Based on these questions, disabilities can be of two types: Congenital diseases are structural or functional abnormalities that arise during intrauterine life. These diseases, sometimes known as birth abnormalities, congenital anomalies, or congenital malformations, arise during pregnancy and can be detected before, during, or after birth.

Acquired - An acquired handicap is one that occurred during the person's lifetime, such as as a result of an accident or chronic illnesses, rather than one that the person was born with. Acquired impairments can be visible or invisible.

According to our findings, whereas congenital disabilities account for 0.8% of all defects, as people grow older, they are more likely to develop disabilities as a result of two key factors: accidents and chronic illnesses. For the sake of our analysis, we will focus on chronic illnesses.

**Increasing Risk with Age:** There's a clear trend showing that heart disease prevalence increases with age. The median prevalence, as well as the interquartile range, expand noticeably for older age groups, indicating both higher median prevalence and greater variability in prevalence rates among these populations.
**Youthful Resilience: **The younger age groups (18-24, 20-24) show considerably lower heart disease prevalence rates, with tighter interquartile ranges suggesting more consistent (lower) prevalence across various states or demographics within these groups.
**Variability in Middle-Aged to Elderly Populations:** The age groups of 45-64, 65+, and 75+ not only have higher prevalence rates but also greater variability. This suggests that while age is a significant risk factor, other factors (such as lifestyle, comorbidities, or access to healthcare) might influence the risk within these age groups, contributing to the wider spread of data.

The examination of key risk factors contributing to heart disease. By focusing on the latest available year in our dataset, we aim to offer a snapshot of the current state of these risk factors among the U.S. adult population

```python
# Selecting a representative year for consistency - Let's use the latest year available in the dataset
latest_year = data['Year'].max()

# Filtering data for key risk factors in the latest year available
risk_factors = [
    'Prevalence of current smoking among US adults (18+) (Percentage); BRFSS',
    'Prevalence of hypertension among US adults (18+) (Percentage); BRFSS',
    'Prevalence of physical inactivity among US adults (18+) (Percentage); BRFSS',
    'Prevalence of obesity among US adults (20+) (Percentage); BRFSS'
]

risk_factor_data = data[data['Indicator'].isin(risk_factors) & (data['Year'] == latest_year)]

# Calculating the average prevalence of each risk factor
average_prevalences = risk_factor_data.groupby('Indicator')['Data_Value'].mean().reset_index()

# Creating the bar chart
plt.figure(figsize=(12, 8))
sns.barplot(x='Data_Value', y='Indicator', data=average_prevalences, palette='coolwarm')
plt.title(f'Average Prevalence of Key Risk Factors in {latest_year}')
plt.xlabel('Average Prevalence (%)')
plt.ylabel('Risk Factor')
plt.tight_layout()
plt.show()
```

The Bar Chart Comparing Average Prevalence of Key Risk Factors in the Latest Year highlights the prevalence of critical factors contributing to heart disease: smoking, hypertension, physical inactivity, and obesity. This visualization serves as a powerful component of our narrative, especially in the conflict section, by demonstrating the scale of public health challenges that need to be addressed to combat heart disease effectively.

```python
# Filter the dataset for additional lifestyle factors: Smoking, Obesity, and High Cholesterol
smoking_data = latest_data[latest_data['Measure'] == 'Current smoking among adults aged >=18 years']
obesity_data = latest_data[latest_data['Measure'] == 'Obesity among adults aged >=18 years']
high_cholesterol_data = latest_data[latest_data['Measure'] == 'High cholesterol among adults aged >=18 years who have been screened in the past 5 years']

# Aggregate these factors by state and calculate average prevalence
states = coronary_heart_disease_data['StateDesc'].unique()

# Initialize a DataFrame to hold aggregated data
lifestyle_factors_df = pd.DataFrame({'State': states})

# Function to aggregate and append data for a given factor
def aggregate_and_append(factor_data, column_name):
    averages = factor_data.groupby('StateDesc')['Data_Value'].mean().reset_index()
    averages.columns = ['State', column_name]
    return averages

# Aggregating data for each factor
smoking_averages = aggregate_and_append(smoking_data, 'Smoking_Prevalence')
obesity_averages = aggregate_and_append(obesity_data, 'Obesity_Prevalence')
high_cholesterol_averages = aggregate_and_append(high_cholesterol_data, 'High_Cholesterol_Prevalence')

# Merging the aggregated data into one DataFrame
lifestyle_factors_df = lifestyle_factors_df.merge(smoking_averages, on='State', how='left')\
                                           .merge(obesity_averages, on='State', how='left')\
                                           .merge(high_cholesterol_averages, on='State', how='left')

# Display the prepared data for visualization
lifestyle_factors_df.head()
```
